```python
import matplotlib.pyplot as plt
```

## Underfitting:
In an underfitting model, both the training and validation errors would be high because the model fails to capture the underlying pattern of the data. The plot would look something like this:

```python
epochs = list(range(1,11))
train_loss = [2.2, 2.1, 2.0, 1.9, 1.8, 1.7, 1.6, 1.5, 1.5, 1.4]
val_loss = [2.3, 2.2, 2.1, 2.0, 1.9, 1.8, 1.7, 1.7, 1.6, 1.6]

plt.plot(epochs, train_loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Underfitting')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

## Overfitting:
In an overfitting model, the training error would be very low because the model is essentially memorizing the training data. However, the validation error would be high because the model fails to generalize to unseen data. The plot would look something like this:

```python
train_loss = [2.2, 1.8, 1.3, 1.0, 0.7, 0.4, 0.3, 0.2, 0.1, 0.1]
val_loss = [2.1, 2.0, 2.0, 2.1, 2.2, 2.3, 2.4, 2.5, 2.6, 2.7]

plt.plot(epochs, train_loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Overfitting')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

## Good fit:
In a well-fitted model, the training error would be low and the validation error would be similar to the training error because the model is able to generalize well to unseen data. The plot would look something like this:


```python
train_loss = [2.2, 1.8, 1.3, 1.0, 0.8, 0.6, 0.5, 0.4, 0.4, 0.3]
val_loss = [2.1, 1.9, 1.5, 1.2, 1.0, 0.9, 0.7, 0.6, 0.5, 0.5]

plt.plot(epochs, train_loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Good Fit')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
